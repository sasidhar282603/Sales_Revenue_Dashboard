```python
# ----------------------------
# Sales & Revenue Analysis Dashboard (Fully Online)
# ----------------------------

import pandas as pd
import plotly.express as px
from io import StringIO

# Step 1: Load sample data from a string
csv_data = """Date,Product,Quantity,UnitPrice
2026-03-01,Product A,10,50
2026-03-01,Product B,5,100
2026-03-02,Product A,7,50
2026-03-02,Product C,12,70
2026-03-03,Product B,8,100
2026-03-03,Product C,5,70
2026-03-04,Product A,6,50
2026-03-04,Product B,10,100
"""

data = pd.read_csv(StringIO(csv_data))
data['Date'] = pd.to_datetime(data['Date'])

# Step 2: Display first few rows
print("First 5 rows of data:")
display(data.head())

# Step 3: Calculate KPIs
total_sales = data['Quantity'].sum()
total_revenue = (data['Quantity'] * data['UnitPrice']).sum()
top_products = data.groupby('Product')['Quantity'].sum().sort_values(ascending=False).head(5)

print(f"\n✅ Total Sales: {total_sales}")
print(f"✅ Total Revenue: ${total_revenue}")
print("\n✅ Top Products:\n", top_products)

# Step 4: Revenue Trend Chart
revenue_trend = data.groupby('Date')['UnitPrice'].sum().reset_index()
fig1 = px.line(revenue_trend, x='Date', y='UnitPrice', title='Revenue Trend Over Time')
fig1.show()

# Step 5: Top Products Chart
fig2 = px.bar(top_products, x=top_products.index, y='Quantity', title='Top Performing Products')
fig2.show()
```
